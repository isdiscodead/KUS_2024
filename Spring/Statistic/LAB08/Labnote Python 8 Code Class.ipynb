```python
# 2.c. Use CLT to randomly draw samples of size “nSample”, up to “nSampleSets” times. 
# E.g. if nSample=10 and nSampleSets=100, we would randomly draw a set of 10 observations 100 times
def CLT(myData, nSample, nSampleSets):
    sampleMean = np.zeros((nSampleSets, 1))
    for i in range(nSampleSets-1):
        x = np.random.choice(len(myData), nSample)
        sampleMean[i] = np.mean(myData[x])
    
    # Provide your answers for (2.c.i), (2.c.ii)
    actVarSampleMean = np.var(sampleMean)
    theorVarSampleMean = np.var(myData) / nSample
    
    print('Empirical sample variance of means is', actVarSampleMean, 'and theoretical sample variance of means is', theorVarSampleMean)
    plt.hist(sampleMean)
```

```python
# Provide your answer for (2.c.iii)
CLT(df.RESACRE, 10, 1000)
CLT(df.RESACRE, 10, 100)
```
