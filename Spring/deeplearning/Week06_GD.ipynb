 ### implement simple gradient descent with Python

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([0.0, 1.0, 2.0])
y = np.array([3.0, 3.5, 5.5])

W = 0 # weight
b = 0 # bias

lrate = 0.01 # learning rate
epochs = 1000 # number of epochs
n = float(len(X)) # number of training data
```

```python
# gradient descent
for i in range(epochs):
    y_pred = W*X + b # prediction

    # gradient descent 수식
    dW = (2/n) * sum(X * (y_pred-y))
    db = (2/n) * sum(y_pred-y)

    W = W - lrate * dW # update weight
    b = b - lrate * db # update bias
```

```python
# print weight and bias.
print (W, b)

# make a prediciton.
y_pred = W*X + b

# plot training data
plt.scatter(X, y)

# plot linear prediction model
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red')
plt.show()
```

#### OTHER VALUES

```python
W = 100 # weight
b = -1 # bias

lrate = 0.1 # learning rate
epochs = 1000 # number of epochs
n = float(len(X)) # number of training data

# gradient descent
for i in range(epochs):
    y_pred = W*X + b # prediction
    dW = (2/n) * sum(X * (y_pred-y))
    db = (2/n) * sum(y_pred-y)
    W = W - lrate * dW # update weight
    b = b - lrate * db # update bias

# print weight and bias.
print (W, b)

# make a prediciton.
y_pred = W*X + b

# plot training data
plt.scatter(X, y)

# plot linear prediction model
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red')
plt.show()
    
```

###  implement linear regression with sklearn I

```python
import matplotlib.pylab as plt
from sklearn import linear_model

reg = linear_model.LinearRegression()

X = [[174], [152], [138], [128], [186]]
y = [71, 55, 46, 38, 88]
```

```python
reg.fit(X, y) # train
print(reg.predict([[165]]))

# scatter plot training data
plt.scatter(X, y, color='black')

# predict the weight value using X of training data
y_pred = reg.predict(X)

# plot a line using X of training data and predicted data
# it shows the linear model
plt.plot(X, y_pred, color='blue', linewidth=3)
plt.show()
```

### implement linear regression with sklearn II

```python
from sklearn import datasets

# load training data
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
 
# select BMI feature and make it into 2-d array. 
# BMI feature index is 2
diabetes_X_new = diabetes_X[:, np.newaxis, 2]

# split training and test data set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(diabetes_X_new, diabetes_y,
test_size=0.1, random_state=0)
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)

# predict with test data.
y_pred = regr.predict(X_test)

# compare result with predicted and real values.
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```
